```python
import pytesseract
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/istockphoto-1484776063-612x612.jpg')
```

```python
image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Convert to RGB (for pytesseract)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Convert to RGB (for pytesseract)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# OCR with Tesseract
text = pytesseract.image_to_string(rgb_image)
print("Detected Text:\n", text)
```

```python
#Display image
plt.imshow(rgb_image)
plt.axis('off')
plt.title("Tesseract OCR Output")
plt.show()
```

```python
# Step 2: Initialize PaddleOCR
from paddleocr import PaddleOCR, draw_ocr

ocr = PaddleOCR(use_angle_cls=True, lang='en') # Initialize PaddleOCR

# Step 3: Run OCR
results = ocr.ocr(image, cls=True)

# Step 4: Print results
for line in results[0]:
    text = line[1][0]
    confidence = line[1][1]
    print(f"{text} (Confidence: {confidence:.2f})")

# Optional: Draw and show results
boxes = [elements[0] for elements in results[0]]
txts = [elements[1][0] for elements in results[0]]
scores = [elements[1][1] for elements in results[0]]

from PIL import Image

image_draw = draw_ocr(image, boxes, txts, scores) # Removed font_path
image_draw = Image.fromarray(image_draw)

plt.imshow(image_draw)
plt.axis('off')
plt.show()
```
