```python
import cv2
```

```python
import matplotlib.pyplot as plt
# read the images
image1 = cv2.imread('/content/image1.jpg')
image2 = cv2.imread('/content/image2.jpg')
```

```python
# Convert BGR to RGB for correct coloring
img1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
# Show both
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img1_rgb)
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img2_rgb)
plt.title('Image 2')
plt.axis('off')

plt.show()
```

```python
# convert images to greyscale
# Convert to grayscale
gray1 = cv2.cvtColor(img1_rgb, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2_rgb, cv2.COLOR_BGR2GRAY)

# Display both grayscale images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray1, cmap='gray')
plt.title('Grayscale Image 1')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray2, cmap='gray')
plt.title('Grayscale Image 2')
plt.axis('off')

plt.show()
```
